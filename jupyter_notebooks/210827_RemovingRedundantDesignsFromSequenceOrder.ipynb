```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import sem
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")

tip_sumstats = (tips.groupby(["day", "sex", "smoker"])
                     .total_bill
                     .agg(["mean", sem])
                     .reset_index())
```
