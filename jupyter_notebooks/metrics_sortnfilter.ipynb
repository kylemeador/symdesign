Importing weights as dict

Now I will import data from `pose_average_metrics_df.csv`. Here, I will have 2 rows for headers:

```python
metrics_file = 'pose_average_metrics_df.csv'

data_df = pd.read_csv(metrics_file, index_col=0, header=[0,1])
data_df
```

```python
if collect_weights:
    weights_dict = SDUtils.query_weights(data_df.columns.to_list())
else:
    weights_dict = {'contact_count': 0.3, 'fsp_energy': 0.5, 'int_separation': 0.1, 'buns_hpol_total':0.1}
```

```python
# Creating df from weights_dict:
wt_dict_df = pd.DataFrame(weights_dict, index=['weight',])
col_list = list(wt_dict_df.columns)
col_list
```

```python
# Creating an empty df called `working_df`, which I will populate with data of those columns that I'm interested in
working_df = pd.DataFrame(index=data_df.index)
```

```python
for col in col_list:
    newdata_df = data_df.loc[:, data_df.columns.get_level_values(1) == col]
    col_names = newdata_df.columns
    for item in col_names:
        working_df[item] = data_df[item]
```

```python
# keeping multiheader consistent
working_df.columns = pd.MultiIndex.from_tuples(working_df.columns, names=['protocol','metric'])
working_df
```

```python
# Importing filter_key
fil_key = pd.read_csv('filter_and_sort_df.csv', index_col=0, nrows=3, header=0)
```

```python
display(fil_key)
```

```python
col_l = list(working_df.columns)
```

```python
ranked_df = pd.DataFrame(index=data_df.index)
```

```python
s_size = len(working_df.index)
step = 1 / ((s_size) - 1)
```

```python
for col in col_l:
    sort_dir = fil_key.loc['sort', col]
    function = fil_key.loc['function', col]
    if sort_dir == 'max' and function == 'sort':
        ranked_df[col] = 1-(((working_df[col].rank(method='dense'))-1)*step)
    elif sort_dir == 'min' and function == 'sort':
        ranked_df[col] = 1-(((working_df[col].rank(method='dense', ascending=False))-1)*step)
    elif sort_dir == 'max' and function == 'normalize':
        max_val = working_df[col].max()
        if max_val < 0:
            ranked_df[col] = max_val/working_df[col]
        else:
            ranked_df[col] = working_df[col]/max_val
    elif sort_dir == 'min' and function == 'normalize':
        min_val = working_df[col].min()
        if min_val < 0:
            ranked_df[col] = working_df[col]/min_val
        else:
            ranked_df[col] = min_val/working_df[col]
```

```python
ranked_df.columns = pd.MultiIndex.from_tuples(ranked_df.columns, names=['protocol','metric'])
ranked_df
```
